## Fandango Exploratory Data Analysis

## Fandango Scores Vs. All Sites

```python
df = pd.merge(fandango, all_sites, on='FILM', how='inner')
```

```python
df.head()
```

```python
df.info()
```

```python
import numpy as np
```

```python
df['RT_Norm'] = np.round(df['RottenTomatoes']/20, 1)
df['RTU_Norm'] = np.round(df['RottenTomatoes_User']/20, 1)
df['MT_Norm'] = np.round(df['Metacritic']/20, 1)
df['MTU_Norm'] = np.round(df['Metacritic_User']/2, 1)
df['IMDB_Norm'] = np.round(df['IMDB']/2, 1)
```

```python
df.head()
```

```python
df.columns
```

```python
norm_scores = df[['STARS', 'RATING', 'RT_Norm', 'RTU_Norm', 'MT_Norm', 'MTU_Norm', 'IMDB_Norm']]
```

```python
norm_scores.head()
```

### Comparing Distribution of Scores Across Sites

```python
def move_legend(ax, new_loc, **kws):
    old_legend = ax.legend_
    handles = old_legend.legendHandles
    labels = [t.get_text() for t in old_legend.get_texts()]
    title = old_legend.get_title().get_text()
    ax.legend(handles, labels, loc=new_loc, title=title, **kws)
```

```python
fig, ax = plt.subplots(figsize=(12, 4), dpi=200)

sns.kdeplot(data=norm_scores, clip=(0, 5), shade=True)

move_legend(ax, "upper left")
```

```python
fig, ax = plt.subplots(figsize=(12, 4), dpi=200)

sns.kdeplot(data=norm_scores[['RT_Norm', 'STARS']], clip=(0, 5), shade=True)

move_legend(ax, "upper left")
```

```python
plt.figure(figsize=(16, 8))
sns.histplot(norm_scores, bins=40)
```

### How are the worst movies rated across all platforms?

```python
sns.clustermap(norm_scores, cmap='magma', col_cluster=False)
```

```python
norm_films = df[['FILM', 'STARS', 'RATING', 'RT_Norm', 'RTU_Norm', 'MT_Norm', 'MTU_Norm', 'IMDB_Norm']]
```

```python
worst_films = norm_films.nsmallest(10, 'RT_Norm')
```

```python
fig, ax = plt.subplots(figsize=(12, 4), dpi=200)

sns.kdeplot(data=worst_films, clip=(0, 5), shade=True)

move_legend(ax, "upper right")
```
